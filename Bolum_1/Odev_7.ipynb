```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
```

```python
my_data = pd.read_csv("desktop/Bootcamp Ödevler/WorldCups.csv", index_col=0)
my_data
```

```python
plt.figure(figsize=(40,5))
plt.subplot(1,2,1)
plt.bar(my_data["Country"], my_data["Winner"], width=0.50, color="red")
plt.title("Dünya Kupası Şampiyonları")
plt.xlabel("Ev Sahipleri")
plt.ylabel("Şampiyonlar")
plt.show()

plt.figure(figsize=(40,5))
plt.subplot(1,2,2)
plt.bar(my_data["Country"], my_data["Runners-Up"], width=0.50, color="yellow")
plt.xlabel("Ev Sahipleri")
plt.ylabel("Finalistler")
plt.title("Dünya Kupası Finalistleri")

plt.show()
```
