```python
#30 yaşından küçük tüm yolcuların kurtulma oranını (survival ratio) hesaplayın.
#kurtulma oranını cinsiyete göre ayrı ayrı hesaplayın.
```

```python
new_df = df[df["Survived"]==1]
a = len(new_df[new_df["Age"]<30]) #30 yaşından küçük olup hayatta kalanların sayısı
b = len(df[df["Age"]<30]) #Toplam 30 yaşından küçük kişi sayısı
survival_ratio= a/b
print("30 yaşından küçük olanların kurtulma oranı= %{:.2f}".format((survival_ratio)*100))


new_df_sex = df[df["Sex"]=="male"] #erkekler
a = len(new_df_sex[new_df_sex["Survived"]==1]) #erkeklerden hayatta kalanlar
b = len(new_df_sex) #toplam erkek sayısı
survival_ratio=a/b*100
print("Erkeklerde hayatta kalma oranı= %{:.2f}".format(survival_ratio))


new_df_sex1 = df[df["Sex"]=="female"] #kadınlar
c = len(new_df_sex1[new_df_sex1["Survived"]==1]) #kadınlardan hayatta kalanlar
d = len(new_df_sex1) #toplam kadın sayısı
survival_ratio=c/d*100
print("Kadınlarda hayatta kalma oranı= %{:.2f}".format(survival_ratio))



    
    
        
```
